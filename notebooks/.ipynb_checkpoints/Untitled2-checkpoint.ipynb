```python
# Fitting a Potential for Each Molecule
# STANDALONE INTEGRATION FOR PG
from sklearn.metrics import r2_score
for molecule in sys_params.keys():
    n = 110
    samp_map = {}
    
    # Import needed data
    try:
        Vpts = np.load('{}/v.npy'.format(sys_params[molecule]['name']))[0:]
        xpts = np.load('{}/xsph.npy'.format(sys_params[molecule]['name']))[0:]
    except FileNotFoundError:
        print("No sampling data for {molecule}.".format(molecule=molecule))
        continue
    
    samp_map['Vpts'] = Vpts
    # First we want to recreate the grid we sampled, and map sample numbers onto angles
    xyz,wts = numgrid.angular_grid(n)
    resol = 2*np.sqrt(np.pi/n)
    chi = np.linspace(0,2*np.pi, int(np.floor(2*np.pi/resol)))[:-1]
    samp_map['Np'] = n
    samp_map['Nq'] = len(chi)
    XYZ = np.array([np.array(coord) for coord in xyz])
    X,Y,Z = XYZ[:,0], XYZ[:,1], XYZ[:,2]
    theta,phi = np.arctan2(np.sqrt(np.power(X,2)+np.power(Y,2)),Z),\
                        np.arctan2(Y,X)%(2*np.pi)
    allcoords = np.array(list(sorted(zip(theta,phi,wts),
                key=lambda t: t[0]))) # angle combinations for spherical grid sorted by theta
    ind = np.lexsort((allcoords[:,1], allcoords[:,0]))
    allcoords = np.array([np.array([allcoords[i,0], allcoords[i,1], allcoords[i,2]]) for i in ind])
    count = 0
    for ch in chi:
        for th,ph,wt in allcoords:
            samp_map[count] = {}
            samp_map[count]['x'] = [ph, th, ch]
            samp_map[count]['w'] = wt
            samp_map[count]['V'] = Vpts[count]
            count += 1
    samp_map['max_count'] = count
            
    print("Fitting potential for {molecule}".format(molecule=molecule))
    LMAX = 0;
    converged = False;
    Lmax_array = [];
    R2_array = [];
    while not converged:    
        # Build a matrix of Wigner D matrix elements
        Dbasis = []
        for pt,v in zip(xpts,Vpts):
            R = quaternionic.array.from_euler_angles(pt[1], pt[0], pt[2])
            wigner = spherical.Wigner(LMAX);
            D = wigner.D(R);
            Dbasis.append(D);
        Dbasis = np.array(Dbasis);
        U,S,V = np.linalg.svd(Dbasis, full_matrices=False)
        ahat = np.matmul( np.linalg.pinv(np.matmul(U, np.matmul( np.diag(S), V))), np.array(Vpts))

        # Check the fit
        predicted = np.zeros(len(Vpts), dtype=np.complex128);
        for i,pt in enumerate(xpts):
            R = quaternionic.array.from_euler_angles(pt[1], pt[0], pt[2]);
            wigner = spherical.Wigner(LMAX);
            D = wigner.D(R);
            predicted[i] += np.dot(ahat, D);
        Lmax_array.append(LMAX);
        R2_array.append(r2_score(Vpts, np.real(predicted)));
        if np.abs(R2_array[-1] - 1) < 0.02:
            converged = True;
            print("Converged at LMAX = {}".format(LMAX))
            print(R2_array[-1])
            break;
        LMAX += 1;
        #print(R2_array[-1], R2r_array[-1])
        print(r2_score(Vpts, np.real(predicted)))
    sys_params[molecule]['v'] = Vpts
    sys_params[molecule]['lmax'] = LMAX
    sys_params[molecule]['ahat'] = ahat
    sys_params[molecule]['samp_map'] = samp_map
    
    mol_dir = os.path.join(save_dir, sys_params[molecule]['name'])
    with open(os.path.join(mol_dir, 'a.txt'), 'w') as f:
        for a in ahat:
            if a is not ahat[-1]:
                f.write('({},{})\n'.format(np.real(a), np.imag(a)))
            else:
                f.write('({},{})'.format(np.real(a), np.imag(a)))
    np.savetxt(os.path.join(mol_dir, 'lmax.txt'), np.array([LMAX]))

def integrate(fpts, samp_map):
    assert len(fpts) == samp_map['max_count']
    integral = 0
    for count,fpt in enumerate(fpts):
        integral += 8*np.pi*np.pi/samp_map['Nq'] * samp_map[count]['w']*fpt
    return integral
```

```python
## T h e r m o d y n a m i c s ##
k = 3.166811e-6
kB = 1.380649e-23   # J/K
R = 0.002 # kcal/mol.K
h = 6.62607004e-34 # Js
c = 3e8 # m/s
Rj = 0.00831 # kJ/mol
HARTREE2KCALMOL = 627.5
HARTREE2KJMOL = 2625.5
HBAR = 1.054571726e-34
NA = 6.02214179e23
EHARTREE = 4.35974434e-18
AMU = 1.660538921e-27
HBAR1 = HBAR / EHARTREE
HBAR2 = HBAR * 1e20 / AMU
H1 = h / EHARTREE
Na = 6.022e23

thermo = {}
for molecule in sys_params.keys():
    print(molecule)
    methods = ['THO',  'TC', 'TPG', 'TVar',
               'PGHO',  'PGC', 'PGPG', 'PGVar',
               'HOHO','HOC', 'HOPG', 'HOVar']
    thermo[molecule] = {'THO' : {'TS' :  0, 'H' : 0, 'G' : 0, 'zpe'  :  0},
                        'TC' : {'TS' :  0, 'H' : 0, 'G' : 0,'zpe'  :  0},
                        'TPG' : {'TS' :  0, 'H' : 0, 'G' : 0,'zpe'  :  0},
                        'TVar' : {'TS' :  0,
                                  'H' : 0,'G' : 0, 'zpe'  :  0},
                       'PGHO' :  {'TS' :  0, 'H' : 0, 'G' : 0,'zpe'  :  0},
                        'PGC' : {'TS' :  0, 'H' : 0, 'G' : 0,'zpe'  :  0},
                        'PGPG' : {'TS' :  0, 'H' : 0, 'G' : 0,'zpe'  :  0},
                        'PGVar' : {'TS' :  0,
                                   'H' : 0, 'G' : 0,'zpe'  :  0},
                       'HOHO'  : {'TS' :  0, 'H' : 0,'G' : 0, 'zpe'  :  0},
                        'HOC' : {'TS' :  0, 'H' : 0,'G' : 0, 'zpe'  :  0},
                        'HOPG' : {'TS' :  0, 'H' :0,'G' : 0, 'zpe'  :  0},
                        'HOVar' : {'TS' :  0,
                                   'H' : 0, 'G' : 0,'zpe'  :  0}}
    #for method in methods:
    #    thermo[molecule][method]['H'] += sys_params[molecule]['ΔE']
    print("***************************************************************************")
    print("***   C O M P U T I N G   T H E R M O D Y N A M I C S   O F   {}   ***".format(molecule))
    print("***************************************************************************\n")
    T = sys_params[molecule]['T']
    sym = sys_params[molecule]['sym']
    mass = sys_params[molecule]['m'] * AMU
    beta = 1/k/T
    ### Rotational Thermodynamics - Gas ###
    Ix,Iy,Iz = sys_params[molecule]['I']
    Qfr = 8*np.pi*np.pi/sym * np.sqrt(np.power(beta*HBAR1*HBAR2*2*np.pi, -3) * Ix*Iy*Iz)
    Ufr = 1.5*Rj*T
    #Uft = 1.5*Rj*T
    Uft = 2.5*Rj*T
    Afr = -Rj*T*np.log(Qfr)
    Sfr = (Ufr - Afr)/T * 1000
    p = 101325 # Pa
    Vol = kB*T / (p) # m^{-3}
    pV = Rj*T
    Λ = np.sqrt((2*np.pi*HBAR*HBAR)/(mass*kB*T))
    Qft = Vol/np.power(Λ,3.0)
    Sft = 1000*Rj*(np.log(Vol*np.power(Λ,-3.0)) + 2.5)
    print("TRANS ENERGY: ", 5/2*Rj*T, "[j/mol]")
    print("ALTERNATE: ", Sft/1000*T - Rj*T*np.log(Qft))

    print("G A S   P H A S E :")
    print("Rotational Partition Function\t\t{}".format(np.around(Qfr,3)))
    print("Rotational Energy\t\t\t{} kJ/mol".format(np.around(Ufr,4)))
    print("Rotational Entropy\t\t\t{} J/mol.K".format(np.around(Sfr,3)))
    print("Translational Partition Function\t{}".format(np.around(Qft,3)))
    print("Translational Energy\t\t\t{} kJ/mol".format(np.around(Uft,4)))
    print("Translational Entropy\t\t\t{} J/mol.K".format(np.around(Sft,3)))
    print("pV\t\t\t\t\t{} kJ/mol".format(np.around(pV,3)))
    for method in methods:
        thermo[molecule][method]['TS'] -= -T*(Sfr+Sft)/1000
        thermo[molecule][method]['H'] -= (Ufr+Uft+pV)
        thermo[molecule][method]['G'] -= (Ufr+Uft+pV)-T*(Sfr+Sft)/1000
        thermo[molecule][method]['zpe'] -= 0
    
    beta = 1/k/T   # in 1/Hartree
    BETA = 1/kB/T
    #  harmonic frequency-related rotational thermodynamics
    rot_vs = sys_params[molecule]['w']*100*c
    #print(rot_vs)
    E0s = 0.5*H1*rot_vs
    zpe_ho = np.sum(E0s)*HARTREE2KJMOL
    Uho = np.sum([E0s/np.tanh(beta1*E0s) for beta1 in beta], axis=1)*HARTREE2KJMOL
    Ucho = np.array([np.sum([1/beta1 for i in range(len(rot_vs))]) for beta1 in beta])*HARTREE2KJMOL
    print(Uho, Ucho)
    Qho = np.array([np.prod(np.power(2*np.sinh(beta1*E0s),-1.0)) for beta1 in beta])
    Qcho = np.array([np.prod(np.power(H1*beta1*rot_vs, -1.0)) for beta1 in beta])
    Fho = -Rj*T*np.log(Qho)
    Fcho = -Rj*T*np.log(Qcho)
    Sho = 1000*(Uho - Fho)/T
    #SHO = (-kB*np.log(1-np.exp(-BETA*h*rot_vs)) + h*rot_vs*100*c/T*np.exp(-BETA*h*rot_vs)*\
    #       (1-np.exp(-BETA*h*rot_vs)))/4184*Na*1000
    Scho = 1000*(Ucho - Fcho)/T
    
    samp_map = sys_params[molecule]['samp_map']
    vpts = samp_map['Vpts']
    Qv = np.array([integrate(np.exp(-beta1*vpts), samp_map) for beta1 in beta])
    num = np.array([integrate(vpts*np.exp(-beta1*vpts), samp_map) for beta1 in beta])
    U = HARTREE2KJMOL*num/Qv + 1.5*Rj*T
    Qt = np.array([np.power(2*np.pi*beta1*HBAR1*HBAR2, -1.5) * np.power(Ix*Iy*Iz, 0.5) for beta1 in beta])
    Q = Qt*Qv/sym
    F = -Rj*T*np.log(Q)
    S = (U-F)/T*1000 # J/mol.K
    
    print("\nA D S O R B E D   P H A S E :")
    print(" - H a r m o n i c    R o t a t i o n - - - - -")
    print("Rotational Partition Function\t\t{}".format(np.around(Qho, 3)))
    print("Rotational ZPE\t\t\t\t{} kJ/mol".format(np.around(zpe_ho,4)))
    print("Rotational Energy\t\t\t{} kJ/mol".format(np.around(Uho,4)))
    print("Rotational Entropy\t\t\t{} J/mol.K".format(np.around(Sho,3)))
    print("Class. Rotational Entropy\t\t{} J/mol.K".format(np.around(Scho,3)))
    print("- - - - - - - - - - - - - - - - - - - - - - - -\n")
    for method in methods:
        if method[-2:] == 'HO':
            thermo[molecule][method]['TS'] += -T*Sho/1000
            thermo[molecule][method]['H']  += Uho
            thermo[molecule][method]['G'] += Uho-T*Sho/1000
            thermo[molecule][method]['zpe']  += zpe_ho
    
    print(" - C l a s s i c a l    R o t a t i o n - - - -")
    print("Rotational Partition Function\t\t{}".format(np.around(Q, 3)))
    print("Rotational Energy\t\t\t{} kJ/mol".format(np.around(U,4)))
    print("Rotational Entropy\t\t\t{} J/mol.K".format(np.around(S,3)))
    print("- - - - - - - - - - - - - - - - - - - - - - - -\n")
    for method in methods:
        if method[-1] == 'C':
            thermo[molecule][method]['TS'] += -T*S/1000
            thermo[molecule][method]['H']  += U
            thermo[molecule][method]['G'] += U - T*S/1000
            thermo[molecule][method]['zpe'] += 0
    
    print(" - P . G    R o t a t i o n - - - - - - - - - -")
    print("Rotational Partition Function\t\t{}".format(np.around(Qho / Qcho * Q, 3)))
    print("Rotational Energy\t\t\t{}  kJ/mol".format(np.around(Uho - Ucho + U,4)))
    print("Rotational Entropy\t\t\t{} J/mol.K".format(np.around(Sho - Scho + S,3)))
    print("- - - - - - - - - - - - - - - - - - - - - - - -\n")
    for method in methods:
        if method[-2:] == 'PG':
            thermo[molecule][method]['TS'] += -T*(Sho - Scho + S)/1000
            thermo[molecule][method]['H']  += Uho - Ucho + U
            thermo[molecule][method]['G'] += Uho - Ucho + U - T*(Sho - Scho + S)/1000
            thermo[molecule][method]['zpe'] += zpe_ho
            
    print(" - V a r i a t i o n a l.  - - - - - - - - - - -")
    for method in methods:
        if method[-3:] == 'Var':
            thermo[molecule][method]['TS'] += - (T * np.array([ 47.71 , 45.98]))/1000
            thermo[molecule][method]['H'] +=    np.array([  10.7  ,  10.167])
            thermo[molecule][method]['G'] += np.array([  10.7  ,  10.167])- (T * np.array([ 47.71 , 45.98]))/1000
            thermo[molecule][method]['zpe']  += 5.38
    
    beta = 1/k/T   # in 1/Hartree
    BETA = 1/kB/T
    #  harmonic frequency-related rotational thermodynamics
    trs_vs = copy.deepcopy(sys_params[molecule]['w_trs'])*100*c
    #print(rot_vs)
    E0s = 0.5*H1*trs_vs
    zpe_ho_t = np.sum(E0s)*HARTREE2KJMOL
    Uho_t = np.array([np.sum(E0s/np.tanh(beta1*E0s))*HARTREE2KJMOL for beta1 in beta])
    Ucho_t = np.array([np.sum([1/beta1 for i in range(len(trs_vs))])*HARTREE2KJMOL for beta1 in beta])
    Qho_t = np.array([np.prod(np.power(2*np.sinh(beta1*E0s),-1.0)) for beta1 in beta])
    Qcho_t = np.array([np.prod(np.power(H1*beta1*trs_vs, -1.0)) for beta1 in beta])
    Fho_t = -Rj*T*np.log(Qho)
    Fcho_t = -Rj*T*np.log(Qcho)
    Sho_t = 1000*(Uho_t - Fho_t)/T
    #SHO = (-kB*np.log(1-np.exp(-BETA*h*rot_vs)) + h*rot_vs*100*c/T*np.exp(-BETA*h*rot_vs)*\
    #       (1-np.exp(-BETA*h*rot_vs)))/4184*Na*1000
    Scho_t = 1000*(Ucho_t - Fcho_t)/T
    
    print(" - H a r m o n i c    T r a n s l a t i o n - -")
    print("Translational Partition Function\t{}".format(np.around(Qho_t, 3)))
    print("Rotational ZPE\t\t\t\t{} kJ/mol".format(np.around(zpe_ho_t,4)))
    print("Translational Energy\t\t\t{}  kJ/mol".format(np.around(Uho_t,4)))
    print("Translational Entropy\t\t\t{} J/mol.K".format(np.around(Sho_t,3)))
    print("- - - - - - - - - - - - - - - - - - - - - - - -\n")
    for method in methods:
        if method[:2] == 'HO':
            thermo[molecule][method]['TS'] += -T*Sho_t/1000
            thermo[molecule][method]['H']  += Uho_t
            thermo[molecule][method]['G']+= Uho_t-T*Sho_t/1000
            thermo[molecule][method]['zpe']+= zpe_ho_t
            
    r_h2o = 1.4 # From literature in Angstrom
    #r_h2o = 1.325
    r_k = sys_params[molecule]['r_k']  # in Angstrom
    rat = np.power(r_h2o/r_k, 3.0)
    print("Ratio is", rat)
    if sys_params[molecule]['zeo'] == 'CHA':
        Vocc = rat * 413.04 * np.power(1e-10, 3) / 2
        Vacc = rat * 413.04 * np.power(1e-10, 3)
    elif sys_params[molecule]['zeo'] == 'MFI':
        Vocc = rat * 511.01 * np.power(1e-10, 3) / 2
        Vacc = rat * 511.01 * np.power(1e-10, 3)
        
    Qtrs_occ = Vocc*np.power(Λ,-3.0)
    Strs_occ = 1000*Rj*(np.log(Vocc*np.power(Λ,-3.0)) + 2.5)
    Ftrs_occ = -Rj*T*np.log(Qtrs_occ)
    Utrs_occ = Ftrs_occ + T*Strs_occ/1000
    pV_occ = 101.325*Vocc*Na
    #pV_occ = 0.17*pV
    Utrs_occ = 1.5*Rj*T
    
    Qtrs_acc = Vacc*np.power(Λ,-3.0)
    Ftrs_acc = -Rj*T*np.log(Qtrs_acc)
    Strs_acc = 1000*Rj*(np.log(0.4*Vacc*np.power(Λ,-3.0)) + 2.5)
    Utrs_acc = Ftrs_acc + T*Strs_acc/1000
    print(" - C l a s s i c a l  T r a n s l a t i o n (Occ.)-")
    print("Translational Partition Function (Occ.)\t{}".format(np.around(Qtrs_occ, 3)))
    print("Translational Energy (Occ.)\t\t{}  kJ/mol".format(np.around(Utrs_occ,4)))
    print("Translational Entropy (Occ.)\t\t{} J/mol.K".format(np.around(Strs_occ,3)))
    print("pV\t\t\t\t\t{} kJ/mol".format(np.around(pV_occ,3)))
    print("- - - - - - - - - - - - - - - - - - - - - - - -\n")
    print(pV_occ)
    for method in methods:
        if method[:1] == 'T':
            thermo[molecule][method]['TS'] += -T*Strs_occ/1000
            thermo[molecule][method]['H']  += Utrs_occ+pV_occ
            thermo[molecule][method]['G'] += Utrs_occ+pV_occ-T*Strs_occ/1000
            thermo[molecule][method]['zpe'] += 0
    
    print(" - P.G.  T r a n s l a t i o n - - -")
    print("Translational Partition Function \t{}".format(np.around(Qho_t/Qcho_t*Qtrs_occ, 3)))
    print("Translational Energy \t\t\t{}  kJ/mol".format(np.around(Uho_t-Ucho_t+Utrs_occ,4)))
    print("Translational Entropy \t\t\t{} J/mol.K".format(np.around(Sho_t-Scho_t+Strs_occ,3)))
    print("- - - - - - - - - - - - - - - - - - - - - - - -\n")
    for method in methods:
        if method[:2] == 'PG':
            thermo[molecule][method]['TS'] += -T*(Sho_t - Scho_t + Strs_occ)/1000
            thermo[molecule][method]['H']  += Uho_t - Ucho_t + Utrs_occ+pV_occ
            thermo[molecule][method]['G'] += Uho_t - Ucho_t + Utrs_occ+pV_occ-T*(Sho_t - Scho_t + Strs_occ)/1000
            thermo[molecule][method]['zpe'] += zpe_ho_t
            
for molecule in molecules:
    for method in methods:
        print(molecule, method, thermo[molecule][method]['H'])
```
