```python
import os
import numpy as np
import numgrid
import quaternionic
import spherical
from sklearn.metrics import r2_score
import cartopy

def integrate(fpts, samp_map):
    assert len(fpts) == samp_map['max_count']
    integral = 0
    for count,fpt in enumerate(fpts):
        integral += 8*np.pi*np.pi/samp_map['Nq'] * samp_map[count]['w']*fpt
    return integral

def get_flmk(fpts, el, m, k, samp_map):
    index = samp_map['wigner_obj'].Dindex(el,m,k)
    Dstar = np.array([samp_map[count]['D'][index] for count in range(samp_map['max_count'])])
    return (2*el+1)/(8*np.pi*np.pi) * integrate(fpts*Dstar, samp_map)

def get_soft(fpts, samp_map):
    fhat = []
    for el in range(samp_map['wigner_lmax']+1):
        for m in np.arange(-el, el+1):
            for k in np.arange(-el, el+1):
                fhat.append(get_flmk(fpts, el, m, k, samp_map))
    f = lambda phi,theta,chi: np.real(np.dot(fhat, 
        samp_map['wigner_D_matrix'](phi,theta,chi)))
    return fhat, f
```

```python
# CLASSICAL THERMO
T = 298.15 #K
T = 303
k = 3.166811e-6
R = 0.002 # kcal/mol
h = 6.62607004e-34 # Js
c = 3e8 # m/s
Rj = 0.00831 # kJ/mol

#. GAS PHASE
Qgas = np.power(2*np.pi*beta*HBAR1*HBAR2, -1.5) * np.power(Ix*Iy*Iz, 0.5) * 8 * np.pi*np.pi /3
Fgas = -R*T*np.log(Qgas)
Ugas = 1.5*R*T
Sgas = (Ugas-Fgas)/T*1000

ngrid = [50, 74, 110,146]
weight = []
coord_dict = {}
for n in ngrid:
    # Import needed data
    xyz,wts = numgrid.angular_grid(n)
    resol = 2*np.sqrt(np.pi/n)
    chi = np.linspace(0,2*np.pi, int(np.floor(2*np.pi/resol)))[:-1]
    print("N AT {} IS {}".format(n, n*len(chi)))
    Vpts = np.load('lebedev{}/v.npy'.format(n))
    pts = np.load('lebedev{}/xsph.npy'.format(n))
    
    system = 'meohmfi74'
    system = 'lebedev110'
    #if not str(n) in system:
    #    continue
    #Vpts = np.load('{}/v.npy'.format(system))
    #pts = np.load('{}/xsph.npy'.format(system))
    
    samp_map = {}
    # First we want to recreate the grid we sampled, and map sample numbers onto angles
    xyz,wts = numgrid.angular_grid(n)
    resol = 2*np.sqrt(np.pi/n)
    chi = np.linspace(0,2*np.pi, int(np.floor(2*np.pi/resol)))[:-1]
    samp_map['Np'] = n
    samp_map['Nq'] = len(chi)
    XYZ = np.array([np.array(coord) for coord in xyz])
    X,Y,Z = XYZ[:,0], XYZ[:,1], XYZ[:,2]
    theta,phi = np.arctan2(np.sqrt(np.power(X,2)+np.power(Y,2)),Z),\
                        np.arctan2(Y,X)%(2*np.pi)
    allcoords = np.array(list(sorted(zip(theta,phi,wts),
                key=lambda t: t[0]))) # angle combinations for spherical grid sorted by theta
    ind = np.lexsort((allcoords[:,1], allcoords[:,0]))
    allcoords = np.array([np.array([allcoords[i,0], allcoords[i,1], allcoords[i,2]]) for i in ind])
    count = 0
    for ch in chi:
        for th,ph,wt in allcoords:
            samp_map[count] = {}
            samp_map[count]['x'] = [ph, th, ch]
            samp_map[count]['w'] = wt
            samp_map[count]['V'] = Vpts[count]
            #print(samp_map[count])
            weight.append(wt)
            count += 1
    samp_map['max_count'] = count

    #num = 0
    #Q = 0
    #for count in range(len(Vpts)):
    #    v = samp_map[count]['V']
    #    #print(v)
    #    num += 8*np.pi*np.pi/samp_map['Nq'] * samp_map[count]['w']*v*np.exp(-beta*v)
    #    Q += 8*np.pi*np.pi/samp_map['Nq'] * samp_map[count]['w']*np.exp(-beta*v)
    beta = 1/k/T
    Q = integrate(np.exp(-beta*Vpts), samp_map)
    num = integrate(Vpts*np.exp(-beta*Vpts), samp_map)
    U = HARTREE2KCALMOL*num/Q + 1.5*R*T
    Qt = np.power(2*np.pi*beta*HBAR1*HBAR2, -1.5) * np.power(Ix*Iy*Iz, 0.5)
    #Qt = np.sqrt(np.pi)/sym * np.sqrt(np.power(beta*HBAR1*HBAR2/2, -3) * Ix*Iy*Iz)
    F = -R*T*np.log(Qt*Q/sym)
    S = (U-F)/T*1000 # cal/mol.K
    #S = (0.885011 + R*T*np.log(149.842))/298.15*1000
    print("NSAMPLES = {}".format(samp_map['max_count']))
    print("RESOLUTION = {} / {}".format(resol, resol*180/np.pi))
    print("*************************************")
    print("*  Classical Thermo for Lebedev {}  *".format(n))
    print("*************************************")
    print("Internal Energy\t", U, (U-Ugas)*4.184)
    print("Entropy\t\t", S, (S-Sgas)*4.184)
    print("Partition Fn\t", Qt*Q/sym)
    
    rot_vs = np.array([169.77, 176.30, 207.01]) # for methane
    rot_vs *= 100*c
    E0s = 0.5*H1*rot_vs
    Uho = np.sum(E0s/np.tanh(beta*E0s))*HARTREE2KCALMOL
    Ucho = np.sum([1/beta for i in range(len(rot_vs))])*HARTREE2KCALMOL
    Qho = np.prod(np.power(2*np.sinh(beta*E0s),-1.0))
    Qcho = np.prod(np.power(H1*beta*rot_vs, -1.0))
    Fho = -R*T*np.log(Qho)
    Fcho = -R*T*np.log(Qcho)
    
    Sho = (Uho - Fho)/T
    Scho = (Ucho - Fcho)/T
    print("**************************************")
    print("* Pitzer-Gwinn Thermo for Lebedev {} *".format(n))
    print("**************************************")
    print("Internal Energy\t", Uho - Ucho + U)
    print("Entropy\t\t", Sho - Scho + S)
    print("Partition Fn\t", Qho/Qcho*Qt*Q/sym)
    print("PG Factor\t", Qho/Qcho)
    print('\n')
    
    print("**************************************")
    print("* Harmonic Thermo for Lebedev {} *".format(n))
    print("**************************************")
    print("Internal Energy\t", Uho)
    print("Entropy\t\t", Sho*1000)
    print("Partition Fn\t", Qho)
    print('\n')
```

```python
# STANDALONE INTEGRATION FOR PG
from sklearn.metrics import r2_score
# Import needed data
system = 'propmfi74' ## PROBLEMS WITH PROP
system = 'ethcha74'
system = 'meohmfi74'
system = 'lebedev110'

Vpts = np.load('lebedev110/v.npy')
pts = np.load('lebedev110/xsph.npy')
Vpts = np.load('{}/v.npy'.format(system))
pts = np.load('{}/xsph.npy'.format(system))
print(len(Vpts))
LMAX = 0;
converged = False;
Lmax_array = [];
R2_array = [];
while not converged:    
    # Build a matrix of Wigner D matrix elements
    Dbasis = []
    for pt,v in zip(pts,Vpts):
        R = quaternionic.array.from_euler_angles(pt[1], pt[0], pt[2])
        wigner = spherical.Wigner(LMAX);
        D = wigner.D(R);
        Dbasis.append(D);
    Dbasis = np.array(Dbasis);
    U,S,V = np.linalg.svd(Dbasis, full_matrices=False)
    ahat = np.matmul( np.linalg.pinv(np.matmul(U, np.matmul( np.diag(S), V))), np.array(Vpts))
    
    # Check the fit
    predicted = np.zeros(len(Vpts), dtype=np.complex128);
    for i,pt in enumerate(pts):
        R = quaternionic.array.from_euler_angles(pt[1], pt[0], pt[2]);
        wigner = spherical.Wigner(LMAX);
        D = wigner.D(R);
        predicted[i] += np.dot(ahat, D);
    Lmax_array.append(LMAX);
    R2_array.append(r2_score(Vpts, np.real(predicted)));
    if np.abs(R2_array[-1] - 1) < 0.05:
        converged = True;
        print("Converged at LMAX = {}".format(LMAX))
        print(R2_array[-1])
        break;
    LMAX += 1;
    #print(R2_array[-1], R2r_array[-1])
    print(r2_score(Vpts, np.real(predicted)))

save_dir = '/Users/lancebettinson/Thesis/umrr/code/hamiltonian-cpp/data/lebedev110'
save_dir = '/Users/lancebettinson/Thesis/umrr/code/hamiltonian-cpp/data/{}'.format(system)
with open(os.path.join(save_dir, 'a.txt'), 'w') as f:
    for a in ahat:
        if a is not ahat[-1]:
            f.write('({},{})\n'.format(np.real(a), np.imag(a)))
        else:
            f.write('({},{})'.format(np.real(a), np.imag(a)))
np.savetxt(os.path.join(save_dir, 'lmax.txt'), np.array([LMAX]))
V = lambda phi,theta,chi: np.real(np.dot(ahat, spherical.Wigner(LMAX).D(quaternionic.array.from_euler_angles(phi,theta,chi))))
np.savetxt(os.path.join(save_dir, 'Eref.txt'), np.array([V(0,0,0)]))
```

```python
# Fitting a dummy potential
# STANDALONE INTEGRATION FOR PG
from sklearn.metrics import r2_score
# Import needed data
system = 'propmfi74' ## PROBLEMS WITH PROP
system = 'ethcha74'
system = 'meohmfi74'

Vpts = np.load('lebedev110/v.npy')[0:]
xpts = np.load('lebedev110/xsph.npy')[0:]

print(len(Vpts), Vpts[0:5], Vpts[-6:-1])
LMAX = 0;
converged = False;
Lmax_array = [];
R2_array = [];
while not converged:    
    # Build a matrix of Wigner D matrix elements
    Dbasis = []
    for pt,v in zip(xpts,Vpts):
        R = quaternionic.array.from_euler_angles(pt[1], pt[0], pt[2])
        wigner = spherical.Wigner(LMAX);
        D = wigner.D(R);
        Dbasis.append(D);
    Dbasis = np.array(Dbasis);
    U,S,V = np.linalg.svd(Dbasis, full_matrices=False)
    ahat = np.matmul( np.linalg.pinv(np.matmul(U, np.matmul( np.diag(S), V))), np.array(Vpts))
    
    # Check the fit
    predicted = np.zeros(len(Vpts), dtype=np.complex128);
    for i,pt in enumerate(xpts):
        R = quaternionic.array.from_euler_angles(pt[1], pt[0], pt[2]);
        wigner = spherical.Wigner(LMAX);
        D = wigner.D(R);
        predicted[i] += np.dot(ahat, D);
    Lmax_array.append(LMAX);
    R2_array.append(r2_score(Vpts, np.real(predicted)));
    if np.abs(R2_array[-1] - 1) < 0.02:
        converged = True;
        print("Converged at LMAX = {}".format(LMAX))
        print(R2_array[-1])
        break;
    LMAX += 1;
    #print(R2_array[-1], R2r_array[-1])
    print(r2_score(Vpts, np.real(predicted)))

V = lambda phi,theta,chi: np.real(np.dot(ahat, spherical.Wigner(LMAX).D(quaternionic.array.from_euler_angles(phi,theta,chi))))
```
